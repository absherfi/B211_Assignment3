```python
import pandas as pd
import numpy as np
import os
```

```python
# Construct the file path
data_dir = r"C:\Users\abbis\Downloads\archive (3)"
file_name = "players_stats_by_season_full_details.csv"
file_path = os.path.join(data_dir, file_name)

# Read the CSV file
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
print(data.head())
```

```python
#calculating the stats

#Field goal percentage
fgm = np.array(data['FGM'])
fga = np.array(data['FGA'])
with np.errstate(divide='ignore', invalid='ignore'):
    fg_percentage = np.where(fga == 0, 0, fgm / fga)

#Three point percentage
three_pm = np.array(data['3PM'])
three_pa = np.array(data['3PA'])
with np.errstate(divide='ignore', invalid='ignore'):
    three_percentage = np.where(three_pa == 0, 0, three_pm / three_pa)

#Free throw percentage
ftm = np.array(data['FTM'])
fta = np.array(data['FTA'])
with np.errstate(divide='ignore', invalid='ignore'):
    ft_percentage = np.where(fta == 0, 0, ftm / fta)

#Points per minute
pts = np.array(data['PTS'])
min = np.array(data['MIN'])
with np.errstate(divide='ignore', invalid='ignore'):
    ppm = np.where(min == 0, 0, pts / min)

#Overall percentage
overall_percentage = (fg_percentage + three_percentage + ft_percentage + ppm)/4

#Blockspergame
blocks = np.array(data['BLK'])
games = np.array(data['GP'])
blockspergame = blocks/games
#Stealspergame
steals = np.array(data['STL'])
stealspergame = steals/games
```

```python
#Create a list of the top 100 players and corresponding season for each of your calculated metrics: 
# field goal accuracy, three point accuracy, free throw accuracy, average points scored per game, 
# overall shooting accuracy, average blocks per game, and average steals per game.

top_100_fg_players = data[['Player', 'Season', 'FGM', 'FGA']].assign(FG_percentage=fg_percentage).sort_values(by='FG_percentage', ascending=False).head(100)
print('The Top 100 Player for Field Goal Accuracy:')
print(top_100_fg_players)
top_100_three_players = data[['Player', 'Season', '3PM', '3PA']].assign(Three_percentage=three_percentage).sort_values(by='Three_percentage', ascending=False).head(100)
print('The Top 100 Player for Three Point Accuracy:')
print(top_100_three_players)
top_100_ft_players = data[['Player', 'Season', 'FTM', 'FTA']].assign(FT_percentage=ft_percentage).sort_values(by='FT_percentage', ascending=False).head(100)
print('The Top 100 Player for Free Throw Accuracy:') 
print(top_100_ft_players)
top_100_ppg_players = data[['Player', 'Season', 'PTS']].sort_values(by='PTS', ascending=False).head(100)
print('The Top 100 Player for Average Points Scored Per Game:')
print(top_100_ppg_players)
top_100_overall_players = data[['Player', 'Season']].assign(Overall=overall_percentage).sort_values(by='Overall', ascending=False).head(100)
print('The Top 100 Player for Overall Shooting Accuracy:')
print(top_100_overall_players)
top_100_blocks_players = data[['Player', 'Season', 'BLK']].sort_values(by='BLK', ascending=False).head(100)
print('The Top 100 Player for Average Blocks Per Game:')
print(top_100_blocks_players)
top_100_steals_players = data[['Player', 'Season', 'STL']].sort_values(by='STL', ascending=False).head(100)
print('The Top 100 Player for Average Steals Per Game:')
print(top_100_steals_players)
```
